```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import calendar
import datetime as dt
```

```python
fcb = pd.read_csv('fcb - fcb.csv')
fcb.columns = ["Season", "División", "Posición", "J", "G", "E", "P", "GF", "GC", "Pts.", "Copa", "Supercopa", "Comp. Europea", "Posición", "SuperCup", "Mundial", "Entrenador", "Presidente"]
fcb = fcb[["Season", "J", "G", "E", "P", "GF", "GC", "Pts."]]

fcb['Season'] = fcb['Season'].str.replace('-..', '', regex=True)
# fcb['Season'].astype('int64')
fcb.set_index('Season', inplace=True)

fcb['Win ratio'] = fcb.G / fcb.J
fcb['G+/G- ratio'] = fcb.GF / (fcb.GF+fcb.GC)
fcb.head(10)
```

```python
rma = pd.read_csv('real_madrid - real_madrid.csv')
rma.columns = ["Season", "División", "Posición", "J", "G", "E", "P", "GF", "GC", "Pts.", "Copa", "Supercopa", "Comp. Europea", "Posición", "SuperCup", "Mundial", "Entrenador", "Presidente"]
rma = rma[["Season", "J", "G", "E", "P", "GF", "GC", "Pts."]]

rma['Season'] = rma['Season'].str.replace('-..', '', regex=True)
# fcb['Season'].astype('int64')
rma.set_index('Season', inplace=True)

rma['Win ratio'] = rma.G / rma.J
rma['G+/G- ratio'] = rma.GF / (rma.GF+rma.GC)
rma.head(10)
```

```python
plt.figure(figsize=(8,6))
plt.subplot(2,2,1)
plt.title('FC Barcelona')
plt.ylabel('Ratio')
fcb['Win ratio'].plot(c='darkred', alpha=0.5, linewidth=3)
fcb['G+/G- ratio'].plot(c = 'darkblue', linewidth=3)
plt.legend()
plt.ylim(0.5,1)

ax=plt.gca()
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.spines["left"].set_visible(False)
ax.spines["bottom"].set_visible(False)

plt.subplot(2,2,2)
plt.title('Real Madrid')
rma['Win ratio'].plot(c='grey', linewidth=3, alpha=0.5)
rma['G+/G- ratio'].plot(c='navy', linewidth=3)
plt.legend()
plt.ylim(0.5,1)

ax=plt.gca()
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.spines["left"].set_visible(False)
ax.spines["bottom"].set_visible(False)

plt.suptitle('Spanish LaLiga study. Comparison between two ratios: goals+/goals- and % Wins')

plt.savefig('foo.png')
```
